```python
import pandas as pd

df = pd.read_excel("./file/전자기기매출액.xlsx")

print(df[::2]) # 모두 가져오는데 2행은 스킵한다는 뜻 0 , 2 , 4 ...
print(df[5:20:2]) # 5 ~ 20 행을 가져오는데 2행 단위로 스킵함 5 , 7 , 9 ...

#df[a:b:c] -> a 와 b 사이를 c 단위로 가져오게 됨

df[::2].to_excel("./file/전자기기매출액_짝수.xlsx")
#to_excel 을 통해서 편집한 데이터프레임을 excel 파일로 저장할 수 있다.
```

```python
df_m = pd.DataFrame()

df_0 = pd.read_excel("./file/2017년_광고비_LG전자.xlsx")
df_1 = pd.read_excel("./file/2017년_광고비_삼성전자.xlsx")
df_2 = pd.read_excel("./file/2017년_광고비_현대자동차.xlsx")

df_0.set_index("date", inplace=True)
df_1.set_index("date", inplace=True)
df_2.set_index("date", inplace=True)
#set_index 는 해당 요소에 따라서 index를 다시 세팅하고 재정렬 한다.

df_m["total"] = pd.concat([df_0["total"],df_1["total"],df_2["total"]], ignore_index = True)
#concat -> pandas dataframe row / col 붙이는 용도 axis = 0 , 1 에 따라서 row , col 모두 가능

df_m.to_excel("./file/전체광고비.xlsx")

df_m
```

```python
ex = pd.ExcelWriter("./file/장사시설현황_합쳐짐.xlsx")
for i in range(2017,2021):
    df_z = pd.read_excel("./file/장사시설현황_2017년_2020년/{}년 장사시설 현황/전국장사시설현황.xlsx".format(i)
                        ,sheet_name="장례식장 시설정보")
    #sheet_name 인자를 추가하면 엑셀파일에서 원하는 sheet 만 불러올 수 있다.
    df_z[["시설명", "주소"]].to_excel(ex , sheet_name="{}년".format(i))
ex.save()

#pandas는 excel 파일을 잘 불러오기는 하지만 쓰는 것은 조금 부족하다.
#그래서 pd.ExcelWriter()을 사용하면 pandas에 엑셀 기능을 더해 사용할 수 있다.

#사용법을 보면 pd.ExcelWriter()로 객체 하나 만들고 거기에 to_excel로 넣는다.
#이때 sheet_name을 정해 sheet 구분을 해서 저장할 수 도 있다.

#모든 입력작업이 끝나면 위애서 만든 객체를 객체.save() 로 저장해주면 된다.
#저장 위치는 처음에 객체를 만들때 정해놓은 위치로 저장이 된다.    
```
